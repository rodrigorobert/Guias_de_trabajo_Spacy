```python
import spacy
```

```python
nlp = spacy.load('es_core_news_sm')
```

```python
documento = nlp(u'estoy esperando a mi mamá en mi casa')
documento
```

```python
# comando para saber información de una palabra
    # .text
    # .pos_
    # .tag
    # .lemma_
    
print(documento[4].text)
print(documento[4].pos_)
print(documento[4].tag_)
print(documento[4].lemma_)
```

```python
documento[4].tag_
```

```python
#para analizar cada palabra hacemos un iteralble

for palabra in documento:
    print(f'{palabra.text} --> {palabra.pos_}')
```

```python
documento2 = nlp(u'yo hago muchos cursos en udemy')

documento2[1].tag_
documento2[1].lemma_
```

```python
# contar cuantas veces aparece algo en una texto

documento3 = nlp(u'MUY LEJOS, MÁS ALLÁ DE LAS MONTAÑAS DE PALABRAS, ALEJADOS DE LOS PAÍSES DE LAS VOCALES Y LAS CONSONANTES, VIVEN LOS TEXTOS SIMULADOS')

for palabra in documento3:
    print(f'{palabra.text:{10}} {palabra.pos_:{10}}')
```

```python
contador = documento3.count_by(spacy.attrs.POS)
contador
```

```python
    #para saber qué es cada número
documento3.vocab[90].text
```

```python
for i,v in sorted(contador.items()):
    print(f'{i} {documento.vocab[i].text} {v}')
```

```python
#representacion gráfica del documento

from spacy import displacy
```

```python
documento4 = nlp(u'Antes de la fiesta de Pascua, sabiendo Jesús que había llegado su hora de pasar de este mundo al Padre, él, que había amado a los suyos que quedaban en el mundo, los amó hasta el fin.')
```

```python
displacy.render(documento4, style='dep', jupyter=True)
```

```python
documento5 = nlp(u'el niño come la manzana')

displacy.render(documento5, style='dep', jupyter=True)
```

```python
#sumar una variable con más opciones de graficación

opciones = {'distance':80, 'compact':'True', 'color':'green', 'bg':'lightgrey', 'font':'Times'}

displacy.render(documento5, style='dep', jupyter=True, options=opciones)
```

```python
#nuevamente, para tokenizar

documento6 = nlp(u'Primera frase. Segunda frase. Tercera frase.')

oraciones = list(documento6.sents)

for i in oraciones:
    print(i)
```

```python
displacy.serve(oraciones, style='dep', options=opciones)
```
